```python
#import dependancies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create plot
plt.plot(x_axis, y_axis)
```

```python
#create plot with object oriented programming 
fig, ax=plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# create the plot
plt.plot(x_axis,y_axis)
plt.show()
```

```python
#creat the plot and add a label for the legend
plt.plot(x_axis,y_axis, label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit between 0 and 45
plt.ylim(0,45)
#create a title
plt.title("pyBer Fare by Month")
#Add the legend

plt.legend()
```

```python
plt.plot(x_axis,y_axis, marker="*", color="blue",linewidth=2,label='Boston')
#create x and y axes labels
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit between 0 and 45
plt.ylim(0,45)
#create a title
plt.title("PyBer Fare by Month")
#add a gid
plt.grid()
#add a legend
plt.legend()
```

```python
plt.plot(x_axis,y_axis, marker="D", color="Green",linewidth=2,label='Boston')
#create x and y axes labels
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit between 0 and 45
plt.ylim(0,45)
#create a title
plt.title("PyBer Fare by Month")
#add a gid
plt.grid()
#add a legend
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis,'o')
```

```python
plt.scatter(x_axis,y_axis)
```

```python
plt.scatter(y_axis,x_axis,c="red",label="Chicago")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
#create bubble charts
plt.scatter(x_axis,y_axis, s=y_axis)
```

```python
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#object oriented scatter
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#object oriented bubble chart
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis, s=y_axis)
```

```python
plt.scatter( y_axis,x_axis, s = [i * 5 for i in y_axis], c="#87cefa", alpha=.8, edgecolors="black", linewidth=2, label="Boston")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%')
plt.show()
```
